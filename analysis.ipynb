```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

import keras as k
from keras.layers import Embedding, Dense, Flatten, Dropout
from keras.models import Sequential
```

```python
#raw_file = pd.read_csv('matches.csv')
raw = pd.read_csv('matches_new_sort.csv')
```

```python
#raw_file.head()
```

```python
# 10 X 120
# raw = raw_file.drop(columns=['start_time'])
raw.columns
```

```python
#raw['radiant_win'] = raw['radiant_win'].astype(int)
target = raw['radiant_win']
raw.drop(columns=['radiant_win'], inplace=True)
raw
```

```python
x_train, x_test, y_train, y_test = train_test_split(raw, target, test_size=.1, random_state=0, shuffle = True)
```

```python
model = Sequential()
model.add(Embedding(121, 11, input_length=10))
model.add(Flatten())
model.add(Dense(1024, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(1, activation='sigmoid'))
model.compile('adam', 'binary_crossentropy', metrics=['acc'])
model.summary()
```

```python
model.fit(x_train, y_train, epochs=5, batch_size=16)
```

```python
pred = model.predict_classes(x_test)
```

```python
print(classification_report(y_test, pred))
```
